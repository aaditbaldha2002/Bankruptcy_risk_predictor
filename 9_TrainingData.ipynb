```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

def get_high_corr_pairs(df, threshold=0.95):
    corr_matrix = df.corr().abs()
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    high_corr_pairs = [
        (row, col, upper_tri.loc[row, col])
        for row in upper_tri.index
        for col in upper_tri.columns
        if upper_tri.loc[row, col] > threshold
    ]
    high_corr_df = pd.DataFrame(high_corr_pairs, columns=["Feature_1", "Feature_2", "Correlation"])
    high_corr_df['AbsCorr'] = high_corr_df['Correlation'].abs()
    return high_corr_df.sort_values(by='AbsCorr', ascending=False)

def filter_unique_pairs(df_corr):
    used_features = set()
    final_pairs = []
    for _, row in df_corr.iterrows():
        f1, f2 = row['Feature_1'], row['Feature_2']
        if f1 not in used_features and f2 not in used_features:
            final_pairs.append(row)
            used_features.update([f1, f2])
    return pd.DataFrame(final_pairs).drop(columns='AbsCorr')

def apply_pca_pairs(df, pairs_df):
    df_out = df.copy()
    new_feature_names = []

    for _, row in pairs_df.iterrows():
        f1, f2 = row['Feature_1'], row['Feature_2']
        data_pair = df_out[[f1, f2]].dropna().values

        if data_pair.shape[0] == 0:
            continue

        # Center data
        data_centered = data_pair - data_pair.mean(axis=0)

        # Apply PCA
        pca = PCA(n_components=1)
        pc1 = pca.fit_transform(data_centered).flatten()

        new_col = f"PCA_{f1}_{f2}"
        df_out[new_col] = np.nan
        df_out.loc[df_out[[f1, f2]].dropna().index, new_col] = pc1
        new_feature_names.append(new_col)

        # Drop original columns
        df_out.drop(columns=[f1, f2], inplace=True)

    return df_out, new_feature_names

def iterative_pca_reduction(df, threshold=0.95, verbose=False):
    df_iter = df.copy()
    iteration = 1
    all_new_features = []

    while True:
        high_corr_df = get_high_corr_pairs(df_iter, threshold=threshold)
        if high_corr_df.empty:
            if verbose:
                print("No more highly correlated feature pairs found.")
            break

        unique_pairs = filter_unique_pairs(high_corr_df)

        if unique_pairs.empty:
            if verbose:
                print("No unique high correlation pairs left.")
            break

        if verbose:
            print(f"\nIteration {iteration}:")
            print(unique_pairs)

        df_iter, new_features = apply_pca_pairs(df_iter, unique_pairs)
        all_new_features.extend(new_features)
        iteration += 1

    return df_iter, all_new_features


final_df, pca_features = iterative_pca_reduction(dataset, threshold=0.90, verbose=True)
final_df=final_df.reset_index(drop=True)

final_df.head(1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer
from tqdm import tqdm  # Optional: for progress tracking

# --------------------------
# STEP 0: Load or Simulate Data
# --------------------------

# Simulate DataFrame with 66 features and 10,000 rows
np.random.seed(42)


# --------------------------
# STEP 1: Plot individual distributions
# --------------------------

def plot_individual_distributions(df, title_prefix="Original Distribution", save=False, folder_path="./plots"):
    for col in tqdm(df.columns, desc=title_prefix):
        plt.figure(figsize=(6, 4))
        sns.histplot(df[col], kde=True, bins=30)
        plt.xlim(df[col].min(), df[col].max())  # Ensures full range is displayed
        plt.title(f"{title_prefix} - {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.tight_layout()

        if save:
            import os
            os.makedirs(folder_path, exist_ok=True)
            plt.savefig(f"{folder_path}/{title_prefix.replace(' ', '_')}_{col}.png")
            plt.close()
        else:
            plt.show()

# Plot original distributions
plot_individual_distributions(final_df, title_prefix="Original Distribution")

# --------------------------
# STEP 2: Apply Yeo-Johnson Gaussianization
# --------------------------

pt = PowerTransformer(method='yeo-johnson', standardize=True)
transformed_data = pt.fit_transform(final_df)
transformed_df = pd.DataFrame(transformed_data, columns=final_df.columns)

# --------------------------
# STEP 3: Plot transformed distributions
# --------------------------

print(transformed_df.head())
plot_individual_distributions(transformed_df, title_prefix="Transformed Distribution")
```
