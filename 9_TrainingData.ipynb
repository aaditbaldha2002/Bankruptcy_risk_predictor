```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

def get_corr_pairs(df, threshold_low=0.85, threshold_high=0.95):
    corr_matrix = df.corr().abs()
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    drop_pairs = []
    pca_pairs = []

    for row in upper_tri.index:
        for col in upper_tri.columns:
            corr = upper_tri.loc[row, col]
            if pd.notna(corr):
                if corr > threshold_high:
                    drop_pairs.append((row, col, corr))
                elif corr > threshold_low:
                    pca_pairs.append((row, col, corr))

    drop_df = pd.DataFrame(drop_pairs, columns=["Feature_1", "Feature_2", "Correlation"])
    pca_df = pd.DataFrame(pca_pairs, columns=["Feature_1", "Feature_2", "Correlation"])
    return drop_df.sort_values(by='Correlation', ascending=False), pca_df.sort_values(by='Correlation', ascending=False)

def filter_unique_pairs(df_corr):
    used = set()
    final = []
    for _, row in df_corr.iterrows():
        f1, f2 = row['Feature_1'], row['Feature_2']
        if f1 not in used and f2 not in used:
            final.append(row)
            used.update([f1, f2])
    return pd.DataFrame(final)

def apply_pca(df, pairs_df):
    df_out = df.copy()
    new_cols = []

    for _, row in pairs_df.iterrows():
        f1, f2 = row['Feature_1'], row['Feature_2']
        if f1 not in df_out.columns or f2 not in df_out.columns:
            continue

        data = df_out[[f1, f2]].dropna().values
        if data.shape[0] == 0:
            continue

        data -= data.mean(axis=0)
        pca = PCA(n_components=1)
        pc1 = pca.fit_transform(data).flatten()

        new_col = f"PCA_{f1}_{f2}"
        df_out[new_col] = np.nan
        df_out.loc[df_out[[f1, f2]].dropna().index, new_col] = pc1
        df_out.drop(columns=[f1, f2], inplace=True)
        new_cols.append(new_col)

    return df_out, new_cols


def apply_drops(df, drop_df):
    df_out = df.copy()
    to_drop = set()
    for _, row in drop_df.iterrows():
        f1, f2 = row['Feature_1'], row['Feature_2']
        if f1 not in to_drop and f2 not in to_drop:
            to_drop.add(f2)  # drop second feature by default
    df_out.drop(columns=list(to_drop), inplace=True, errors='ignore')
    return df_out

def hybrid_iterative_reduction(df, thresh_low=0.8, thresh_high=0.98, verbose=False):
    df_iter = df.copy()
    iteration = 1
    all_pca_cols = []

    while True:
        drop_df, pca_df = get_corr_pairs(df_iter, threshold_low=thresh_low, threshold_high=thresh_high)

        if drop_df.empty and pca_df.empty:
            if verbose:
                print("No more correlated features to process.")
            break

        if verbose:
            print(f"\n--- Iteration {iteration} ---")
            if not drop_df.empty:
                print("Drop Pairs:")
                print(drop_df)
            if not pca_df.empty:
                print("PCA Pairs:")
                print(pca_df)

        if not drop_df.empty:
            df_iter = apply_drops(df_iter, drop_df)

        if not pca_df.empty:
            unique_pca_df = filter_unique_pairs(pca_df)
            df_iter, new_pca_cols = apply_pca(df_iter, unique_pca_df)
            all_pca_cols.extend(new_pca_cols)

        iteration += 1

    return df_iter.reset_index(drop=True), all_pca_cols

final_df, pca_features = hybrid_iterative_reduction(dataset, thresh_low=0.85, thresh_high=0.95, verbose=True)
final_df.head(1)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

inertia_list = []
silhouette_scores = []

K_range = range(3, 7)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(transformed_df)
    
    # Record inertia (SSE)
    inertia_list.append(kmeans.inertia_)
    
    # Record silhouette score
    score = silhouette_score(transformed_df, kmeans.labels_)
    silhouette_scores.append(score)
```

```python
# --- Step 1: Clustering the dataset ---

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Standardize the features before clustering (very important for KMeans)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(dataset)

# 2. Define number of clusters
team_size = 3  # <<== change this according to your real team size
k = max(team_size, min(2 * team_size, 6))  # k between team_size and 2*team_size

# 3. Perform KMeans clustering
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# 4. Add cluster labels to dataset
dataset['cluster_id'] = cluster_labels

# 5. Also keep in mind isBankrupt for each index
dataset['isBankrupt'] = isBankrupt.values

# --- Step 2: Report number of companies and bankruptcy proportion in each cluster ---

report = dataset.groupby('cluster_id').agg(
    total_companies=('isBankrupt', 'count'),
    bankrupt_companies=('isBankrupt', 'sum')
).reset_index()

report['bankrupt_proportion'] = report['bankrupt_companies'] / report['total_companies']

print("\n=== Cluster Report ===")
print(report)

# --- Step 3: Visualize and find cluster characteristics ---

# 3.1 PCA projection for visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=cluster_labels, palette='Set2')
plt.title('PCA Projection of Companies with Cluster Labels')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.show()

# 3.2 Boxplot of a few important features across clusters
important_features = dataset.drop(columns=['cluster_id', 'isBankrupt']).columns[:5]  # pick first 5 features for now

for feature in important_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=dataset, x='cluster_id', y=feature)
    plt.title(f'Feature: {feature} by Cluster')
    plt.show()
```

```python
# --- 3.3.1: Build a model to predict cluster_id (classification) ---

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Features and labels
X = dataset.drop(columns=['cluster_id', 'isBankrupt'])
y = dataset['cluster_id']

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2)

# Random Forest Classifier (can overfit slightly as allowed)
cluster_predictor = RandomForestClassifier(n_estimators=200, random_state=42)
cluster_predictor.fit(X_train, y_train)

# Predict and evaluate
y_pred = cluster_predictor.predict(X_test)

print("\n=== Cluster Prediction Report ===")
print(classification_report(y_test, y_pred))
print("\n=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred))

# Important features
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': cluster_predictor.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("\n=== Top Important Features for Cluster Prediction ===")
print(feature_importances.head(10))
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.ensemble import StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.exceptions import FitFailedWarning
import warnings

# Suppress FitFailedWarnings
warnings.filterwarnings("ignore", category=FitFailedWarning)

# Set up
clusters = dataset['cluster_id'].unique()
final_results = []

for cluster in clusters:
    print(f"\n=== Working on Cluster {cluster} ===")

    # Extract data for this cluster
    cluster_data = dataset[dataset['cluster_id'] == cluster]

    # Features and labels
    X_cluster = cluster_data.drop(columns=['cluster_id', 'isBankrupt'])
    y_cluster = cluster_data['isBankrupt']

    # If no bankrupt companies OR no non-bankrupt companies, skip
    if len(y_cluster.unique()) < 2:
        constant_value = y_cluster.iloc[0]
        print(f"Cluster {cluster} has only one class ({constant_value}). Assigning constant model (always predict {constant_value}).")
        continue

    # Base models (non-parametric)
    base_models = [
        ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
        ('knn', KNeighborsClassifier(n_neighbors=5)),
        ('dt', DecisionTreeClassifier(max_depth=5, random_state=42)),
    ]

    # Meta-model (no use_label_encoder needed)
    meta_model = xgb.XGBClassifier(eval_metric='logloss', verbosity=0, random_state=42)

    # Adjust number of folds if needed
    if y_cluster.value_counts().min() < 5:
        cv_folds = 3
    else:
        cv_folds = 5

    # Stacking classifier
    stack_model = StackingClassifier(
        estimators=base_models,
        final_estimator=meta_model,
        cv=StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42),
        passthrough=True,
        n_jobs=-1
    )

    # Train
    stack_model.fit(X_cluster, y_cluster)

    # Cross-validation predictions
    y_pred_cv = cross_val_predict(stack_model, X_cluster, y_cluster, cv=cv_folds, method='predict')

    # Evaluate
    acc = np.mean(y_pred_cv == y_cluster)
    cm = confusion_matrix(y_cluster, y_pred_cv)

    print(f"Accuracy for Cluster {cluster}: {acc:.4f}")
    print(f"Confusion Matrix:\n{cm}")

    # Save results
    final_results.append({
        'Cluster': cluster,
        'Accuracy': acc,
        'Confusion Matrix': cm
    })
```
